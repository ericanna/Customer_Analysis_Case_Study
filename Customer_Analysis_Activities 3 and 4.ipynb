```python
##Importing the necessary libraries: Pandas, Numpy, Matplotlib and Seaborn.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
##Importing a csv file with the workable data do Python.

clean_df = pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
clean_df
clean_df.info()
```

```python
# Activity: Get the numeric data into dataframe called numerical and categorical columns in a dataframe called categoricals.

## Using the np.number and object methods to segregate the numerical data types and categorical data types respectively.

numerical_df = clean_df.select_dtypes(include=np.number)
categorical_df = clean_df.select_dtypes(include=object)
numerical_df

```

```python
# Activity: Show a plot of the total number of responses.

sns.histplot(categorical_df["response"])
```

```python
# Activity: Show a plot of the response by the sales channel.

## Used "hue" since we are now working with two variables.

sns.countplot(x='sales_channel', hue='response', data=categorical_df)
plt.xlabel("Sales Channel")
plt.ylabel("Count")
plt.legend(title="Response")
plt.title("Response by Sales Channel")
```

```python
# Activity: Show a plot of the response by the total claim amount.

## First transformed the total_claim_amount, that comes in thousands of diferent values, in ranges of value. To do...
## ... that, need to check the actual range of the values, by applying the .min and .max methods.

print(numerical_df["total_claim_amount"].min())
print(numerical_df["total_claim_amount"].max())


def ranges_claim(value):
    if value >= 0 and value <= 100:
        return 0
    elif value > 100 and value <= 500:
        return 1
    elif value > 500 and value <= 1000:
        return 2
    elif value > 1000 and value <= 1500:
        return 3
    elif value > 1500 and value <= 2000:
        return 4
    else:
        return 5
```

```python
clean_df["total_claim_amount"] = list(map(ranges_claim, clean_df["total_claim_amount"]))

## Then used "hue", since working with two variables.

sns.countplot(x='total_claim_amount', hue='response', data=clean_df)
plt.xlabel("Total Claim Amount")
plt.ylabel("Count")
plt.legend(title="Response", loc="upper right",fontsize=12)
plt.title("Response by Total Claim Amount")
plt.xticks(ticks=np.arange(6),labels=["$0 - $100,00", "$100,01 - $500,00","$500,01 - $1000,00","$1000,01 - $1500,00","$1500,01 - $2000,00", "More then $2000,00"],rotation=90) 
```

```python
# Activity: Show a plot of the response by income.

## Same as above, transforming the income, that comes in thousands of diferent values, in ranges of value, after...
## ...checking the range of values.

print(numerical_df["income"].min())
print(numerical_df["income"].max())

def ranges_income(value):
    if value > 0 and value <= 15000:
        return 0
    elif value > 15000 and value <= 30000:
        return 1
    elif value > 30000 and value <= 45000:
        return 2
    elif value > 45000 and value <= 60000:
        return 3
    elif value > 60000 and value <= 80000:
        return 4
    else:
        return 5


clean_df["income"] = list(map(ranges_income, clean_df["income"]))
```

```python
## Then created the graphic using "hue", since working with two variables.

sns.countplot(x='income', hue='response', data=clean_df)
plt.xlabel("Income")
plt.ylabel("Count")
plt.legend(title="Response", loc="upper right",fontsize=8)
plt.title("Response by income")
plt.xticks(ticks=np.arange(6),labels=["$0 - $15.000,00", "$15.000,01 - $30.000,00","$30.000,01 - $45.000,00","$45.000,01 - $60.000,00","$60.000,01 - $80.000,00", "More then $80.000,00"],rotation=90) 
```
